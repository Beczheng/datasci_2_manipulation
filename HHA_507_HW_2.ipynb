# Importing packages

```python
from faker import Faker
import numpy as np
import pandas as pd
import polars as pl
import re
import time
```

# Cleaning the dataset

## Removing duplicates

```python
# Removing duplicates in the dataset

df2 = df.drop_duplicates()
df2
```

## Renaming missing values

```python
# Replacing the word "missing" with "NaN" in the dataset

df3 = df2.replace('missing', np.nan)
df3
```

## Identifying missing values

```python
# Identifying missing values in the dataset

df3.isnull()
```

```python
# Identifying missing values in each row in the dataset

df3.isnull().sum(axis=1)
```

## Dropping missing values

```python
# Dropping missing values in the dataset

df4 = df3.dropna()
df4
```

## Renaming other values

```python
# Replacing the word "Female" with "F" under the "Gender" column

df5 = df4.replace('Female', 'F')
df5
```

```python
# Replacing the word "Male" with "M" under the "Gender" column

df6 = df5.replace('Male', 'M')
df6
```

## Removing special characters in values

```python
# Removing the underscore symbol for the data under the "Medication Prescribed" column

df6['Medication Prescribed'] = df6['Medication Prescribed'].str.replace('[^a-zA-Z0-9]',' ')
df6['Medication Prescribed']
```

## Changing column names



```python
# Removing white space and special characters in column names

def clean_column(df):
  def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

  df.columns = [clean_name(col) for col in df.columns]
  return df

df7 = clean_column(df6)
df7
```

```python
# Renaming the column names

df8 = df7.rename(columns = {'patientage' : 'patient_age', 'cityofresidence' : 'city', 'stateofresidence' : 'state', 'hasinsurance' : 'has_insurance', 'visitedlastmonth' : 'visited_last_month', 'paymentmethod' : 'payment_method', 'preferreddoctor' : 'preferred_doctor', 'diseasediagnosed' : 'disease_diagnosed', 'medicationprescribed' : 'medication_prescribed', 'typeofappointment' : 'type_of_appointment', 'averageheartrate' : 'average_hr', 'averagebp' : 'average_bp', 'heightincm': 'height_in_cm', 'weightinkg' : 'weight_in_kg', 'paymentdue' : 'payment_due', 'lastvisitdaysago' : 'last_visit_in_days_ago', 'visitdurationmins' : 'visit_duration_in_mins', 'numberoftests' : 'number_of_tests', 'prescriptioncost' : 'prescription_cost'})
df8
```

```python
# Generating a new column using the current dataset

df8['average_hr'] = df8['average_hr'].astype(int)
df8['average_hr_status'] = np.where((df8['average_hr'] >= 60) & (df8['average_hr'] <= 100), 'Normal', 'Not normal')
df8
```

```python
# Generating a new column using a fake healthcare dataset

fake = Faker()

num_patients = 10

data = {
    'patient_age': [fake.random_int(min=1, max=120) for _ in range(num_patients)],
    'average_hr': [fake.random_int(min=0, max=200) for _ in range(num_patients)],
}

new_data = pd.DataFrame(data)

# Defining thresholds for categorization

threshold_high = 100
threshold_low = 60

# Creating a new column "average_hr_status" based on thresholds

new_data['average_hr_status'] = new_data['average_hr'].apply(lambda x: 'High' if x >= threshold_high else 'Low' if x <= threshold_low else 'Normal')
print(new_data[['patient_age', 'average_hr', 'average_hr_status']].head(5))
```
